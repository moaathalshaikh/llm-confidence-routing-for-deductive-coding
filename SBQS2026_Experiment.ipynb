# SBQS 2026 — Deductive Coding of CI Claims
## Context Granularity Experiment: L1–L5

**Research Question:**  
How does context granularity (bibliographic metadata) affect LLM performance in deductive coding of SE evidence claims?

| Level | Name | Content |
|-------|------|---------|
| L1 | Claim Only | claim text only |
| L2 | Source Context | claim + study title |
| L3 | Methodological Context | claim + title + research method |
| L4 | Empirical Context | claim + title + variables |
| L5 | Full Context | claim + title + method + variables |

**Models:** Claude Haiku 4.5 · DeepSeek Chat · GPT-4o Mini  
**Dataset:** 113 single-label CI claims with human gold standard

## Step 7 — Preview results

```
import pandas as pd

analysis_dir = f'{PROJECT_DIR}/analysis'

print('=== Accuracy by Model × Level ===')
metrics = pd.read_csv(f'{analysis_dir}/metrics_by_model_level.csv')
code_metrics = metrics[metrics['task']=='code']
pivot = code_metrics.groupby(['model','level'])['accuracy'].mean().unstack('level').round(3)
print(pivot)

print('\n=== Cohen Kappa (Model vs Gold) ===')
kappa = pd.read_csv(f'{analysis_dir}/kappa_analysis.csv')
mvg = kappa[(kappa['kappa_type']=='model_vs_gold') & (kappa['task']=='code')]
pivot_k = mvg.groupby(['model_a','level'])['kappa'].mean().unstack('level').round(3)
print(pivot_k)
```
